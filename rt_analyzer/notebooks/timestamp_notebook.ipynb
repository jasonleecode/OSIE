```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
f=open("/home/ivan/repos/nexedi/osie/rt_analyzer/channel1_duration.txt", "r")
lines = f.readlines()
f.close()
lines = [float(x.replace("\n", "")) for x  in lines[:]]
```

```python
d = {}
i = 0
for x in lines:
    d[i] = x
    i += 1
s = pd.Series(d, name='duration')
```

```python
s.plot.bar()
plt.show()
```
